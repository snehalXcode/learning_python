```python
'''Take an integer n and:Check whether it is a prime number
ðŸ“Œ Focus: loop + condition
â›” Donâ€™t use libraries.'''

n = int(input("Enter a number: "))

if n < 2:
    print(f"{n} is not prime.")
else:
    is_prime = True

    for i in range(2, n):
        if n % i == 0:
            is_prime = False
            break

    if is_prime:
        print(f"{n} is a prime number!")
    else:
        print(f"{n} is not a prime number.")
```

```python
'''Given a list:nums = [2, 3, 4, 5, 6, 7, 8] Create a new list containing:only prime numbers
ðŸ“Œ Focus: loops + logic reuse (from Q1)'''
nums = [2, 3, 4, 5, 6, 7, 8]
primes = []

for n in nums:
    if n < 2:
        continue

    is_prime = True
    for i in range(2, n):
        if n % i == 0:
            is_prime = False
            break

    if is_prime:
        primes.append(n)

print("List of primes:", primes)
```

```python
'''Take a string input and:

Count how many uppercase and lowercase letters are present.

Example:

Input: "PyTHon"
Uppercase = 3
Lowercase = 3'''

text = input("Enter the string: ")

upper_count = 0
lower_count = 0

for char in text:
    if char.isupper():
        upper_count += 1
    elif char.islower():
        lower_count += 1

print("Uppercase =", upper_count)
print("Lowercase =", lower_count)
```

```python
'''Take an integer n and:Reverse the number Check whether it is a palindrome
Example:
Input: 121
Output: Palindrome
ðŸ“Œ Focus: while loop + math'''

n = input("Enter: ")

if n == n[::-1]:
    print("Palindrome")
else:
    print("Not a Palindrome")
```
