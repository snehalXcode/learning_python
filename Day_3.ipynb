```python
'''Take an integer input n and:Print all numbers from 1 to n Skip numbers divisible by 3
ðŸ“Œ Focus: loop + continue'''

n = int(input("Enter the number : "))

for i in range (1,n+1):
  if i%3==0:
    continue
  else :
    print(i)
```

```python
'''Given a list:nums = [1, 2, 3, 4, 5] Calculate and print:Sum of all elements Average of the elements
ðŸ“Œ Focus: loop + arithmetic'''

nums = [1,2,3,4,5]
sum = 0
for n in nums:
  sum=n+sum
  avg=sum/len(nums)
print(sum)
print(avg)
```

```python
'''Given a list:nums = [3, 5, 7, 9] Check whether the list is strictly increasing.
Output:"Increasing" or "Not Increasing"
ðŸ“Œ Focus: comparison inside loop'''

nums = [3, 5, 7, 9]
is_increasing = True
for i in range(1, len(nums)):
    if nums[i] <= nums[i - 1]:
        is_increasing = False
        break

if is_increasing:
    print("Increasing")
else:
    print("Not Increasing")
```
