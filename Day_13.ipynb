```python
'''Take an integer n and:Find the sum of digits greater than 5
Example:
Input: 58243
Output: 15   # (8 + 7)
ðŸ“Œ Hint:if digit > 5 â†’ add digit'''

n = input("Enter the number: ")
total_sum = 0

for char in n:
    digit = int(char)
    if digit > 5:
        total_sum += digit

print(total_sum)
```

```python
'''Take an integer n and:Count how many digits are equal to 3
Example:
Input: 23341
Output: 2
ðŸ“Œ Hint:
if digit == 3 â†’ count++'''


n = input("Enter the number: ")
count = 0

for char in n:
    digit = int(char)
    if digit == 3 :
      count = 1+count
print(count)

```

```python
'''Take an integer n and:Create a new number using only odd digits
Example:
Input: 482731
Output: 731
ðŸ“Œ Hint:check odd digit build new number using *10 + digit'''

n = input("Enter the number: ")
new_num = 0

for char in n:
    digit = int(char)
    if digit % 2 != 0:          # If digit is odd
        new_num = new_num * 10 + digit

print(new_num)
```

```python
'''Given a list:nums = [2, 5, 8, 11, 14, 17]
Create a new list with:
numbers that are both odd and greater than 10'''

nums = [2, 5, 8, 11, 14, 17]
new = []

for n in nums:
  if n%2 != 0 and n > 10:
    new.append(n)
print(new)
```
