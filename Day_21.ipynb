```python
'''Given a list:nums = [3, 7, 2, 9, 12, 15, 18]
Create a new list with:numbers that are greater than 5 AND odd'''

nums = [3, 7, 2, 9, 12, 15, 18]
new = []

for n in nums:
  if n > 5 and n % 2 != 0:
    new.append(n)

print(new)
```

```python
'''Given:nums = [10, 21, 30, 41, 50, 61]
Count how many numbers:are not divisible by 5'''

nums = [10, 21, 30, 41, 50, 61]
count = 0

for n in nums :
  if n % 5 != 0 :
    count = 1 + count

print(count)
```

```python
'''Take an integer n and:Reverse the number
Example:
Input: 5824
Output: 4285
(No [::-1] tricks. Use logic.)'''

n = int(input("Enter : "))
reversed_num = 0

while n > 0:
    # 1. Get the last digit
    last_digit = n % 10

    # 2. Build the new number: shift left and add
    reversed_num = (reversed_num * 10) + last_digit

    n = n // 10

print(reversed_num)
```

```python
'''Given a list:nums = [4, 6, 3, 9, 2, 8]
Find the second largest number.'''

nums = [4, 6, 3, 9, 2, 8]

nums.sort()
print(nums[-2])
```
