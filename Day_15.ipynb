```python
'''Take an integer n and:

Count how many digits are even AND greater than 4

Example:

Input: 4827316
Output: 2   # (6, 8)'''

n = input("Enter : ")
count=0

for char in n:
  digit = int(char)
  if digit > 4 and digit % 2 == 0:
    count=1+count
print(count)
```

```python
'''Take an integer n and:Find the sum of digits that are multiples of 3
Example:
Input: 96342
Output: 18   # (9 + 6 + 3)'''

n = input("Enter : ")
sum=0

for char in n:
  digit = int(char)
  if digit % 3 == 0:
    sum=digit+sum
print(sum)
```

```python
'''Take an integer n and:Create a new number using digits that are odd
Example:
Input: 482731
Output: 731'''

n = input("Enter the number : ")
new = 0

for char in n:
  digit = int(char)
  if digit % 2 != 0 :
    new = new * 10  + digit
print(new)
```

```python
'''Given a list:nums = [4, 7, 10, 13, 16, 19]
Create a new list with:numbers that are odd and greater than 10'''

nums = [4, 7, 10, 13, 16, 19]
new=[]

for n in nums:
  if n>10 and n%2 != 0:
    new.append(n)

print(new)
```

```python
'''Take a string input and:Count how many characters are digits
Example:
Input: "ab12c3"
Output: 3'''

n = input("Enter a string: ")
count = 0

for char in n:
    if char.isdigit():
        count += 1

print(count)
```
