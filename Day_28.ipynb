```python
'''Given:nums = [2, 4, 7, 9, 12, 15, 18]
Count how many numbers are:
odd AND divisible by 3'''

nums = [2, 4, 7, 9, 12, 15, 18]
count = 0
for num in nums:
    if num % 3 == 0 and num % 2 != 0:
        count = 1 + count

print(count)
```

```python
'''Given:nums = [3, 6, 2, 9, 5]
Find the sum of all elements except smallest
ðŸ‘‰ WITHOUT using min()'''

nums = [3, 6, 2, 9, 5]
total_sum = 0
smallest = float('inf')

for n in nums:
  if n < smallest:
    smallest = n
  total_sum = n + total_sum

print(total_sum - smallest)
```

```python
'''Take an integer n and:

Find the sum of digits at even positions

Example:

Input: 12345
Positions: 1 2 3 4 5
Digits:    1 2 3 4 5
Even positions â†’ 2, 4
Output: 6'''

n_str = input("Enter number : ")
current_sum = 0

for i in range(len(n_str)):
    # Positions are 1-indexed, so even positions correspond to odd indices (1, 3, 5, ...)
    if (i + 1) % 2 == 0:
        current_sum += int(n_str[i])

print(current_sum)
```

```python
'''Given:
nums = [5, 2, 9, 1, 7]
Find the difference between second largest and second smallest
ðŸ‘‰ WITHOUT sorting'''

nums = [5, 2, 9, 1, 7]
largest = float('-inf')
second_largest = float('-inf')
smallest = float('inf')
second_smallest = float('inf')

for num in nums:
  if num > largest:
    second_largest = largest
    largest = num
  elif num > second_largest and num != largest:
    second_largest = num

  if num < smallest:
    second_smallest = smallest
    smallest = num
  elif num < second_smallest and num != smallest:
    second_smallest = num

print(second_largest - second_smallest)
```
