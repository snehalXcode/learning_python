```python
'''Take an integer n and:

Reverse the number

Check if it is palindrome

Example:

Input: 121
Output: Palindrome'''

n = input("Enter number  : ")
rev = n[::-1]

if n == rev :
  print("Palindrome")
else :
  print(rev)
```

```python
'''Given:nums = [4, 7, 2, 9, 5]
Find:
largest
smallest
difference'''

nums = [4, 7, 2, 9, 5]
largest = 0
smallest = 0

for n in nums :
    if n > largest :
       largest = n
    if n < smallest :
       smallest = n

print(largest)
print(smallest)
print(largest - smallest)
```

```python
'''Given:nums = [3, 6, 7, 9, 10, 12, 15]
Create a new list with:numbers that are divisible by 3 AND odd'''

nums = [3, 6, 7, 9, 10, 12, 15]
new = []

for n in nums :
  if n % 3 == 0 and n % 2 != 0 :
    new.append(n)

print(new)
```

```python
'''Take an integer n and:Find the second largest digit'''

n = input("Enter a number : ")
largest = 0
second = 0

for i in n :
  if int(i) > largest :
    second = largest
    largest = int(i)

  elif int(i) > second and int(i) != largest :
    second = int(i)

print(second)
```
