```python
'''Take an integer n and:Find the factorial of n
Example:
Input: 5
Output: 120
ðŸ“Œ Focus: loop + accumulator'''

n=int(input("Enter : "))
product=1

for i in range(1,n+1):
  product=product*i

print(product)
```

```python
'''Given a list:nums = [10, 15, 20, 25, 30]
Create a new list containing:numbers divisible by 10'''

nums = [10,15,20,25,30]
div = []

for n in nums:
  if n % 10 == 0:
   div.append(n)
print(div)
```

```python
'''Take an integer n and:Check whether it is an Armstrong number
Example:
Input: 153
Output: Armstrong
ðŸ“Œ Focus: while loop + math'''

# Step 1: Get user input
n = int(input("Enter : "))

# Step 2: Store the original number and find the number of digits
temp = n
order = len(str(n)) # Total number of digits
sum_result = 0      # Accumulator

# Step 3: While loop to process each digit
while temp > 0:
    digit = temp % 10          # Extracts the last digit (math: remainder)
    sum_result += digit ** order # Adds digit^order to the sum
    temp //= 10                # Removes the last digit (math: floor division)

# Step 4: Comparison
if n == sum_result:
    print("Armstrong")
else:
    print("Not Armstrong")
```
