```python
'''Take an integer n and:Find the sum of even digits in the number
Example:
Input: 123456
Output: 12   # (2 + 4 + 6)'''

n_str = input("Enter : ")
total_sum = 0

# Iterate through each character in the string
for digit in n_str:
    digit_int = int(digit) # Convert character back to integer for math
    if digit_int % 2 == 0: # Check if the specific digit (not the whole number) is even
        total_sum += digit_int # Add the even digit to the sum

print(total_sum)
```

```python
'''Given a list:

nums = [3, 8, 1, 9, 5]


Find and print:

Smallest element

Largest element

â›” Donâ€™t use min() or max()'''

nums=[3,8,1,9,5]
largest=nums[0]
smallest=nums[0]

for n in nums:
  if n > largest:
    largest = n
  if n < smallest:
    smallest = n
print(largest)
print(smallest)
```

```python
'''Take a string input and:

Count how many words start with a vowel

Example:

Input: "apple is an orange"
Output: 3'''

n = str(input("Enter : "))
count = 0
vowels = "aeiouAEIOU" # Define a string containing all vowels

# Step 1: Split the sentence into a list of words
words = n.split()

# Step 2: Loop through each word in the list
for word in words:
    if word[0] in vowels:
        count += 1

print(count)

```

```python
''''Take an integer n and:

Convert it to binary
(Do it manually using logic, not bin())

Example:

Input: 5
Output: 101


ðŸ“Œ Focus: while loop + remainder'''

n = int(input("Enter : "))
binary = ""

# Handle the case if n is 0
if n == 0:
    binary = "0"

while n > 0:
    remainder = n % 2          # Get the remainder (0 or 1)
    binary = str(remainder) + binary  # Add remainder to the FRONT of the string
    n = n // 2                 # Perform integer division to reduce n

print(binary)
```
