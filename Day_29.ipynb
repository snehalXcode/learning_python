```python
'''Take an integer n and:

Reverse the number

Example:

Input: 1234
Output: 4321'''

n = input("Enter number : ")
rev = n[::-1]
print(rev)
```

```python
'''Take an integer n and:

Find the sum of digits at odd positions

Example:

Input: 12345
Odd positions â†’ 1, 3, 5
Output: 1 + 3 + 5 = 9'''

n_str = input("Enter number : ")
current_sum = 0

for i in range(len(n_str)):
    if (i + 1) % 2 != 0:
        current_sum += int(n_str[i])

print(current_sum)
```

```python
'''Given:

nums = [4, 7, 2, 9, 5]

Find the second smallest number'''

nums = [4, 7, 2, 9, 5]
nums.sort()
second_smallest = nums[1]
print(f"The second smallest number is: {second_smallest}")
```

```python
'''Given:

nums = [3, 8, 1, 9, 5]

Find:

largest

second largest

ðŸ‘‰ WITHOUT using max() or sort()'''

nums = [3, 8, 1, 9, 5]

largest = float('-inf')
second_largest = float('-inf')

for num in nums:
    if num > largest:
        second_largest = largest
        largest = num
    elif num > second_largest and num != largest:
        second_largest = num

print(f"The largest number is: {largest}")
print(f"The second largest number is: {second_largest}")
```
