```python
'''Take an integer n and:

Count how many digits are greater than 2 AND even

Example:

Input: 4827316
Output: 3   # (4, 8, 6)'''

n = input("Enter : ")
count = 0

for char in n:
  digit=int(char)
  if digit > 2 and digit % 2 == 0 :
    count = 1 + count
print(count)
```
