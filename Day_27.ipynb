```python
'''Given:nums = [8, 3, 11, 2, 9, 5]
Find:largest number,smallest number'''

nums = [8, 3, 11, 2, 9, 5]
largest = nums[0]
smallest = nums[0]

for n in nums:
  if n > largest :
    largest = n
  elif n < smallest :
    smallest = n
  else:
    continue

print(f'largest number is {largest}')
print(f'smallest number is {smallest}')
```

```python
'''Take an integer n and:Check whether the number is a palindrome
Example:
Input: 121
Output: Yes'''

n = input("Enter a number :" )
rev = n[::-1]

if n == rev :
  print("Yes")
else :
  print("no")
```

```python
'''Given:
nums = [4, 7, 2, 9, 5] Find the sum of second largest and second smallest numbers'''

nums = [4, 7, 2, 9, 5]
sum = 0
nums.sort()

second_smallest = nums[1]
second_largest = nums[-2]

sum = second_smallest + second_largest
print(sum)
```
