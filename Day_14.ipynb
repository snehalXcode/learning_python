```python
'''Take an integer n and:Count how many digits are less than 5
Example:
Input: 482731
Output: 3   # (4, 2, 3)'''

n = input("Enter the number : ")
count = 0

for char in n :
  digit = int(char)
  if digit < 5 :
    count = 1+count
print(count)
```
