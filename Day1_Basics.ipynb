```python
'''Write a program that:Takes an integer n prints numbers from 1 to n But:
Print "Fizz" if number is divisible by 3
Print "Buzz" if divisible by 5
Print "FizzBuzz" if divisible by both
ðŸ“Œ Focus: ifâ€“elifâ€“else, %'''

n = int(input("Enter number: "))

for i in range(1, n + 1):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
```

```python
'''Q2 Given a list:nums = [2, 4, 6, 8, 10, 3] Check if all elements are even.If yes, print "All Even" else print "Not All Even".
ðŸ“Œ Focus: loops, flags, break'''

nums = [2, 4, 6, 8, 10, 3]

all_even = True   # flag

for num in nums:
    if num % 2 != 0:
        all_even = False
        break

if all_even:
    print("All Even")
else:
    print("Not All Even")
```

```python
'''Q3 Take a string input and:Count vowels Count consonants
Example:
Input: "python"
Output:
Vowels = 1
Consonants = 5
ðŸ“Œ Focus: strings, conditions""'''

text = input("Enter a string: ").lower()
vowels = 0
consonants = 0

for char in text:
    if char.isalpha():  # Check if it's a letter
        if char in "aeiou":
            vowels += 1
        else:
            consonants += 1

print("Vowels =", vowels)
print("Consonants =", consonants)


```

```python
''' Q4 Given a list:arr = [1, 2, 3, 4, 5, 6] Create two lists:One with even numbers One with odd numbers
ðŸ“Œ Focus: lists, logic separation     '''

arr = [1, 2, 3, 4, 5, 6]
even_numbers = []
odd_numbers = []
for num in arr:
    if num % 2 == 0:
        even_numbers.append(num)
    else:
        odd_numbers.append(num)
print(f"Original list: {arr}")
print(f"Even numbers: {even_numbers}")
print(f"Odd numbers: {odd_numbers}")
```

```python
''' Q5 Take an integer n and print this pattern:Input: 4
*
* *
* * *
* * * *
ðŸ“Œ Focus: nested loops, pattern thinking'''

n = int(input("Enter a number: "))

# Outer loop for the number of rows
for i in range(1, n + 1):
    # Inner loop for the number of stars in each row
    for j in range(i):
        print("*", end=" ")
    # New line after each row
    print()
```
