```python
'''Take an integer n and:

Count how many digits are divisible by 3

Example:

Input: 93642
Output: 4   # (9, 3, 6, 4? nope â†’ mistake avoided ðŸ˜„)
Correct digits: 9, 3, 6 â†’ Output: 3'''

n=input("Enter the number  : ")
count=0

for char in n:
  digit = int(char)
  if digit % 3 == 0 :
    count = 1 + count
print(count)
```

```python
'''Take an integer n and:

Find the sum of digits that are prime

ðŸ“Œ Reminder:

prime digits = 2, 3, 5, 7

Example:

Input: 23581
Output: 10   # (2 + 3 + 5)'''

n = input("Enter the number: ")
total_sum = 0

for char in n:
    digit = int(char)
    if digit == 2 or digit == 3 or digit == 5 or digit == 7:
        total_sum = total_sum + digit

print("Sum of prime digits:", total_sum)
```

```python
'''Given a list:

nums = [1, 4, 6, 7, 9, 10, 11]


Create a new list with:

numbers that are prime'''

nums = [1, 4, 6, 7, 9, 10, 11]
new = []  # Correct: Initialize as an empty list

for n in nums:
    if n < 2:
        continue  # Skip 1 and negative numbers as they are not prime

    is_prime = True
    # Check for factors from 2 up to the square root of n
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            is_prime = False
            break

    if is_prime:
        new.append(n)

print(new)  # Output: [7, 11]
```
