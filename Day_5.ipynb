```python
'''Take an integer input n and:

Print numbers from 1 to n

Replace multiples of 4 with "Quad"

ðŸ“Œ Focus: loop + %'''

n=int(input("Enter : "))

for i in range(1,n+1):
  if i%4 == 0:
    print("Quad")
  else:
    print(i)
```

```python
'''âœ… Q2 (Easy)

Given a list:

nums = [5, 10, 15, 20, 25]

Count how many numbers are:

Greater than 15

ðŸ“Œ Focus: loop + condition + counter'''

nums=[5,10,15,20,25]
count=0

for n in nums:
  if n > 15 :
    count=1+count
print(count)

```

```python
'''Take a string input and:

Check whether it is a palindrome or not

Example:

Input: "madam"
Output: Palindrome

ðŸ“Œ Focus: string logic'''

n=str(input("enter :"))
rev=n[::-1]

if n == rev:
  print("Palindrome")
else :
  print("not")
```

```python
'''Given a list:

nums = [4, 6, 8, 9, 10]

Find and print:

First odd number in the list
(Stop once found)

ðŸ“Œ Focus: loop + break'''

nums = [4,6,8,9,10]

for n in nums:
  if n%2 != 0:
    print(n)
    break
```
