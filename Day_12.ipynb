```python
'''Take an integer n and:Find the sum of odd digits
ðŸ“Œ Hint:
digit = n % 10
check digit % 2 != 0
add to sum n //= 10'''

n = int(input("Enter the number: "))
sum_odd = 0

while n > 0:
    digit = n % 10
    if digit % 2 != 0:
        sum_odd = digit+sum_odd
    n //= 10

print(sum_odd)
```

```python
'''Take an integer n and:Check whether the number contains digit 5
Example:
Input: 4821
Output: No

ðŸ“Œ Hint:
Loop through digits
If digit == 5 â†’ stop'''

n = int(input("Enter the number : "))
found = False

while n > 0:
    digit = n % 10
    if digit == 5:
        found = True
        break
    n //= 10

if found:
    print("Yes")
else:
    print("No")
```

```python
'''Take an integer n and:Count how many odd digits are present
ðŸ“Œ Hint:
digit = n % 10
check digit % 2 != 0
counter++
n //= 10'''
n = int(input("Enter the number : "))
count = 0

while n > 0:
    digit = n % 10
    if digit % 2 != 0:
        count = 1 + count
    n //= 10

print(count)
```

```python
'''Given a list:nums = [5, 12, 7, 18, 9, 20]
Create a new list containing:
only numbers that are even
ðŸ“Œ Focus:
loop
condition
append'''

nums = [5, 12, 7, 18, 9, 20]
new = []

for n in nums :
  if n % 2 == 0 :
    new.append(n)
print(new)
```

```python
'''Take an integer n and:Find the sum of even digits
ðŸ“Œ Hint:
digit = n % 10
if digit % 2 == 0 â†’ add to sum
n //= 10
Example:
Input: 24631
Output: 12   # (2 + 4 + 6)'''

n = int(input("Enter the numbers : "))
sum = 0

while n > 0 :
  digit = n % 10
  if digit % 2 == 0:
    sum = digit + sum

    n //=10
print(sum)
```

```python
'''Take an integer n and:Count how many digits are greater than 5
ðŸ“Œ Hint:
extract digit
check digit > 5
increase counter
remove digit'''

n = int(input("Enter the number: "))
count = 0

while n > 0:
    digit = n % 10
    if digit > 5:
        count += 1
    n //= 10

print(count)
```
