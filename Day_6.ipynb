```python
'''Q1 (Easy)

Take an integer input n and:

Print numbers from 1 to n

Skip numbers divisible by 2 or 3

ðŸ“Œ Focus: loop + continue'''


n = int(input("Enter the number: "))
for i in range(1, n + 1):
    if i % 2 == 0 or i % 3 == 0:
        continue
    print(i)
```

```python
'''Given a list:

nums = [12, 5, 8, 20, 7]

Count how many numbers are:

Greater than 10

ðŸ“Œ Focus: loop + counter'''

nums=[12,5,8,20,7]
count=0

for n in nums:
  if n > 10 :
    count=1+count

print(count)
```

```python
'''Take a string input and:

Count how many words are in the string

Example:

Input: "I love python"
Output: 3

ðŸ“Œ Focus: strings + logic'''

n=str(input("Enter the string : "))
words = n.split()
word_count=len(words)
print(word_count)
```

```python
'''Q4 (Medium)

Given a list:

nums = [2, 4, 6, 8, 9]

Check whether the list is:

"All Even"

or "Contains Odd"

ðŸ“Œ Focus: flag + loop'''

nums = [2, 4, 6, 8, 9]
all_even = True

for n in nums:
    if n % 2 != 0:
        all_even = False
        break

if all_even:
    print("All Even")
else:
    print("Contains Odd")
```
