```python
'''Given:nums = [3, 8, 1, 9, 5]
Find the largest number'''

nums = [3,8,1,9,5]
largest = nums[0]

for n in nums :
  while n > 0 :
    if n > largest :
      largest = n
    break
print(largest)
```

```python
'''Given:nums = [7, 2, 10, 4, 1]
Find the smallest number'''

nums = [7, 2, 10, 4, 1]
smallest = nums[0]

for n in nums :
    if n < smallest :
      smallest = n
    break
print(smallest)
```

```python
'''nums = [4, 7, 12, 15, 18, 21]
Create a new list with:
numbers that are divisible by 3 AND greater than 10'''

nums = [4, 7, 12, 15, 18, 21]
new = []

for n in nums  :
  if n % 3 == 0 and n > 10:
    new.append(n)
print(new)
```

```python
'''Take an integer n and:Find the second largest digit
Example:
Input: 5827316
Output: 7'''

n = input("Enter number : ")

digits = sorted(list(set([int(d) for d in n])), reverse=True)

if len(digits) >= 2:
  print(digits[1])
else:
  print("There is no second largest digit.")
```
