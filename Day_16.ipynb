```python
'''Take an integer n and:

Count how many digits are odd

ðŸ“Œ Hint:

digit = n % 10

if digit % 2 != 0 â†’ count++

n //= 10'''

n = int(input("Enter: "))
count = 0

while n > 0:
    digit = n % 10       # Get the last digit
    if digit % 2 != 0:   # Check if it's odd
        count += 1
    n //= 10             # Remove the last digit

print("Odd digits:", count)
```
