```python
'''âœ… Q1 (List â€“ Manual Second Largest)
Given:nums = [4, 7, 2, 9, 5]
Find the second largest number
ðŸ‘‰ WITHOUT using sort() or max()'''

nums = [4, 7, 2, 9, 5]

largest = nums[0]
second_largest = nums[0]

take = 0
while take < len(nums):
    current = nums[take]

    if current > largest:
        second_largest = largest
        largest = current

    elif current > second_largest:
        second_largest = current

    take += 1

print("The second largest is:", second_largest)
```

```python
'''Given:nums = [3, 6, 7, 9, 10, 12, 15]
Create a new list with:numbers that are divisible by 3 AND NOT even'''


nums = [3, 6, 7, 9, 10, 12, 15]
new = []

for n in nums:
  if n % 3 == 0 and n % 2 == 0 :
    new.append(n)

print(new)
```

```python
'''Given:
nums = [4, 7, 2, 9, 5]
Find the sum of all elements except largest'''

nums = [4, 7, 2, 9, 5]

sum = 0
largest = nums[0]

for n in nums:
    sum = sum + n
    if n > largest:
        largest = n

result = sum - largest

print(result)
```
