```python
'''Take an integer n and:Count how many digits it has
ðŸ“Œ Hint:
Use while n > 0
Increase a counter'''

n = int(input("Enter: "))
count = 0

while n > 0:
    count += 1
    n = n // 10

print(count)
```

```python
'''Take an integer n and:Find the sum of digits
ðŸ“Œ Hint:
digit = n % 10
remove digit = n // 10'''

n = int(input("Enter : "))
sum = 0
while n > 0:
   digit = n % 10
   sum = sum + digit
   n = n // 10
print(sum)
```

```python
'''Given a list:nums = [1, 3, 5, 7, 9]
Check whether all elements are odd.'''

nums = [1, 3, 5, 7, 9]
all_odd = True

for n in nums:
    if n % 2 == 0:
        all_odd = False
        break

print(all_odd)
```

```python
'''Take an integer n and:Count how many even digits are present
ðŸ“Œ Hint:
digit = n % 10
check digit % 2 == 0
remove digit â†’ n //= 10'''
n = int(input("Enter  : "))
count = 0

while n > 0 :
  digit = n % 10

  # Check if the digit is even
  if digit % 2 == 0:
    count = count + 1  # Increase the count ONLY if it's even

  n = n // 10

print(f"Total even digits: {count}")
```
