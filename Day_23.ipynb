```python
 '''Given:nums = [3, 8, 11, 14, 17, 20, 23]
Create a new list with:numbers that are odd AND greater than 10'''

nums = [3, 8, 11, 14, 17, 20, 23]
new=[]

for n in nums:
  if n > 10 and n % 2 != 0 :
    new.append(n)

print(new)
```

```python
'''Given:nums = [5, 10, 15, 20, 25, 30]
Count how many numbers:are divisible by 5 AND even'''

nums = [5, 10, 15, 20, 25, 30]
count = 0

for n in nums:
  if n % 5 == 0 and n % 2 == 0:
    count = 1 + count

print(count)
```

```python
'''Given:nums = [4, 7, 2, 9, 5]
Find the difference between largest and smallest number'''

nums = [4, 7, 2, 9, 5]

largest = max(nums)
smallest = min(nums)

difference = largest - smallest

print("Output:", difference)
```
