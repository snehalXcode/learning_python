```python
'''Q2 Take an integer n and:Print sum of numbers from 1 to n
Example:
Input: 5
Output: 15
ðŸ“Œ Focus: loop + accumulator'''

n = int(input("Enter number : "))
sum_total = 0

for i in range(1, n + 1):
    sum_total = i+sum_total

print(sum_total)
```

```python
'''Take a string input and:Print the string in reverse
Example:
Input: "python"
Output: "nohtyp"
ðŸ“Œ Focus: loops OR slicing (your choice)'''

n = input("Enter string: ").lower()

rev = n[::-1]

print(rev)
```

```python
'''Q5 Take an integer n and print:Input: 3
1
12
123
ðŸ“Œ Focus: nested loops (simple)'''

n = int(input("Enter number: "))
for i in range(1, n + 1):
    for j in range(1, i + 1):
        print(j, end="")
    print()
```
