```python
'''Given:nums = [5, 2, 9, 1, 7]
Find the second largest number'''

nums = [5, 2, 9, 1, 7]
largest = nums[0]
second_largest = nums[0]

for n in nums:
    if n > largest:
        second_largest = largest
        largest = n
    elif n > second_largest:
            second_largest = n

print(second_largest)
```

```python
'''nums = [4, 7, 10, 13, 16, 19]
Create a new list with:numbers that are odd AND less than 15'''

nums = [4, 7, 10, 13, 16, 19]
new = []

for n in nums :
    if n % 2 != 0 and n < 15:
        new.append(n)

print(new)
```

```python
'''Take an integer n and:Find the difference between largest and smallest digit
Example:
Input: 5827316
Output: 7   '''

n = input("Enter the number: ")

largest = int(n[0])
smallest = int(n[0])

for char in n:
    digit = int(char)

    if digit > largest:
        largest = digit

    if digit < smallest:
        smallest = digit

print(largest - smallest)
```
