```python
'''Take an integer n and:

Count how many digits are even OR equal to 5

Example:

Input: 5827316
Output: 4   # (8, 2, 6, 5)'''

n = input("Enter : ")
count = 0

for char in n:
  digit=int(char)
  if digit % 2 == 0 or digit ==5:
    count=1+count
print(count)
```

```python
'''Take an integer n and:

Find the sum of digits that are odd AND greater than 3

Example:

Input: 4827316
Output: 10   # (7 + 3)'''

n = input("Enter : ")
sum = 0

for char in n:
  digit=int(char)
  if digit % 2 != 0 and digit >  3:
    sum=digit+sum
print(sum)
```

```python
'''Take an integer n and:

Create a new number using digits that are less than 6

Example:

Input: 5827316
Output: 231'''

n = input("Enter : ")
new = []

for char in n:
    digit = int(char)
    if digit < 6:
        new.append(char)

result = "".join(new)
print(result)
```

```python
'''Given a list:

nums = [2, 4, 5, 7, 8, 11, 14]


Create a new list with:

numbers that are odd OR greater than 10'''

nums = [2, 4, 5, 7, 8, 11, 14]
new=[]

for n in nums :
  if n % 2 != 0 or n > 10 :
    new.append(n)
print(new)
```
