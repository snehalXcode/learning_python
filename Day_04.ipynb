```python
'''Given a list:nums = [10, 20, 30, 40] Calculate and print:Sum of elements Product of elements
ðŸ“Œ Focus: loop + accumulator'''
nums = [10, 20, 30, 40]
sum_total = 0
product = 1

for n in nums:
    sum_total = n+sum_total
    product *= n

print("Sum of elements:", sum_total)
print("Product of elements:", product)
```

```python
'''Take a string input and:Count how many uppercase letters are present
Example:
Input: "PyTHon"
Output: 3
ðŸ“Œ Focus: string methods (isupper())'''

n = str(input("Enter :  "))
count=0
for char in n:
    if char.isupper():
        count += 1
print(count)
```

```python
'''Given a list:nums = [2, 3, 5, 7, 11] Check whether all numbers are prime.
Output:"All Prime" or "Not All Prime"
ðŸ“Œ Focus: nested loops + logic
â›” Donâ€™t use libraries.'''

nums = [2,3,5,7]
all_prime =  True

for n in nums :
  if n < 2 :
    all_prime=False
    break
for i in range(2, n):
        if n % i == 0:
            all_prime = False
            break

if all_prime:
    print("All Prime")
else:
    print("Not All Prime")
```
