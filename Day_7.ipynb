```python
'''Take an integer n and print:
*
* *
* * *
* * * *
üìå Hint:Row number = number of stars'''

n=int(input("Enter : "))

for i in range(1,n+1):
  for j in range(1,i+1):
    print("*",end=" ")
  print()
```

```python
'''Take an integer n and print:
* * * *
* * *
* *
*
üìå Hint:Inner loop decreases'''

n = int(input("Enter : "))

for i in range (n):
  for j in range (n-i):
    print("*",end=" ")
  print()
```

```python
'''Given a list:nums = [3, 6, 9, 12, 15]
Print only numbers divisible by 3.
üìå Focus: loop + condition'''

nums = [3,6,9,12,15]

for n in nums :
  if n % 3 != 0:
    continue
  else:
    print(n)
```

```python
'''Take an integer n and print:
1
2 2
3 3 3
4 4 4 4
üìå Hint:Outer loop controls the number Inner loop controls how many times it prints'''

n = int(input("Enter : "))

for i in range(1,n+1):
  for j in range(1,i+1):
    print(i,end=" ")
  print()
```
