```python
'''Take an integer n and:Count how many digits are odd AND less than 7
Example:
Input: 4827316
Output: 2   # (3, 1)'''

n = input("Enter the number : ")
count=0

for char in n:
  digit=int(char)
  if digit < 7 and digit % 2 != 0:
    count = 1+count
print(count)
```

```python
'''Take an integer n and:Find the sum of digits that are multiples of 4
Example:
Input: 4827316
Output: 12   # (4 + 8)'''

n = input("Enter the number : ")
sum = 0

for char in n:
  digit=int(char)
  if digit % 4 == 0 :
    sum = digit+sum
print(sum)
```

```python
'''Given a list:
nums = [3, 4, 5, 8, 9, 12, 15]
Create a new list with:numbers that are divisible by 3 but NOT by 4'''

nums = [3, 4, 5, 8, 9, 12, 15]
new = []

for n in nums:
  if n % 3 == 0 and n % 4 != 0:
    new.append(n)
print(new)
```
